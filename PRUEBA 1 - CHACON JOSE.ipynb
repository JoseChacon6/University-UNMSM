**EJERCICIO 6**

¿Cuantos números por encima de 20 existen en la siguiente lista? Resuelvalo usando for.

```
lista = [10, 40, 60, 70, 80]
mayor = []
for i in lista:
  if i >20:
    print("{} es mayor que 20".format(i))
    mayor.append(i)
  elif i == 20:
    print("{} es 20".format(i))
  else:
    print("{} es menor que 20".format(i))
print("son {} numeros mayores que 20".format(len(mayor)))
```

**EJERCICIO 7**

Crear un programa que mencione si un número es positivo, negativo o cero.

```
num = [0, -2, 4, 5, 6]

for i in num:
  if i > 0:
    print("{} es un numero positivo".format(i))
  elif i == 0:
    print("{} es igual a 0".format(i))
  else:
    print("{} es un numero negativo".format(i))
```
